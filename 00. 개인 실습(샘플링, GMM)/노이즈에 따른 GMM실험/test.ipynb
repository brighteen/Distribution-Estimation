```python
# 최신 cuML 버전(24.04+)에서는 클래스 이름이 GMM으로 변경되었습니다.
from cuml.cluster import GMM
import cudf
import numpy as np

# 예제 데이터 생성 (GPU에서 처리하기 위해 cudf 데이터프레임 사용)
print("cuDF로 GPU에 예제 데이터를 생성합니다...")
X_gpu = cudf.DataFrame()
X_gpu['col1'] = np.random.rand(100).astype(np.float32)
X_gpu['col2'] = np.random.rand(100).astype(np.float32)

# GMM 객체 생성 (GaussianMixture가 아닌 GMM을 사용)
print("GMM 모델 객체를 생성합니다...")
gmm_model = GMM(n_components=3)

# 모델 학습
print("모델 학습을 시작합니다...")
gmm_model.fit(X_gpu)

print("\n모델 학습이 성공적으로 완료되었습니다!")
print("데이터의 각 샘플에 할당된 클러스터 레이블:")
print(gmm_model.labels_)
```
