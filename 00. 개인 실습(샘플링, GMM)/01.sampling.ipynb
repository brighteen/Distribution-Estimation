```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

# 균등 분포에서 샘플링

```python
# MNIST 이미지 크기와 동일하게 28x28로 설정
height = 28
width = 28

# 28x28 크기의 흑백 이미지를 생성
# np.random.rand(h, w)는 h x w 형태의 2차원 배열을 생성하고,
# 각 요소를 [0, 1) 범위의 균일 분포에서 무작위로 샘플링합니다.
# 채널(channel) 차원이 없으므로 흑백 이미지가 됩니다.
random_grayscale_image = np.random.rand(height, width)

print(f"image의 평균값: {np.mean(random_grayscale_image)}, 표준편차: {np.std(random_grayscale_image)}, shape: {random_grayscale_image.shape}")
```

```python
type(random_grayscale_image)
```

```python
random_grayscale_image
```

```python
# 생성된 흑백 이미지 시각화
plt.imshow(random_grayscale_image, cmap='gray') # 흑백으로 표시하기 위해 cmap='gray' 설정
plt.title("28x28 Grayscale Image from Uniform Distribution")
plt.axis('off')
plt.show()
```

```python
# 1. 균등 분포(Uniform Distribution) 시각화 함수
def visualize_uniform_distribution(height=28, width=28, bins=50):
    """
    균등 분포에서 샘플링하고 히스토그램으로 시각화합니다.
    U(0, 1) 분포를 따르므로 값은 0과 1 사이에 평평하게 분포합니다.
    """
    print("--- 균등 분포 시각화 ---")

    # [0, 1) 범위에서 균등 분포로 데이터 샘플링
    uniform_data = np.random.rand(height, width)

    # 히스토그램 생성
    plt.figure(figsize=(10, 6))
    plt.hist(uniform_data.flatten(), bins=bins, density=True, alpha=0.7, label='Sampled Uniform Data')

    # 이론적인 확률 밀도 함수(PDF) 그리기 (U(0,1)의 PDF는 y=1)
    x = np.linspace(0, 1, 100)
    plt.plot(x, stats.uniform.pdf(x), 'r-', lw=2, label='Ideal Uniform PDF')

    plt.title('Visualization of Uniform Distribution (0, 1)')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
visualize_uniform_distribution()
```

# 가우시안 분포에서 샘플링

```python
random_grayscale_image = np.random.randn(height, width)

print(f"image의 평균값: {np.mean(random_grayscale_image)}, 표준편차: {np.std(random_grayscale_image)}, shape: {random_grayscale_image.shape}")
```

```python
plt.imshow(random_grayscale_image, cmap='gray')
plt.title("28x28 Grayscale Image from Normal Distribution")
plt.axis('off')
plt.show()
```

```python
# 2. 가우시안(정규) 분포 시각화 함수
def visualize_gaussian_distribution(height=28, width=28, bins=50):
    """
    표준 정규 분포에서 샘플링하고 히스토그램으로 시각화합니다.
    N(0, 1) 분포를 따르므로 값은 평균 0을 중심으로 종 모양을 이룹니다.
    """
    print("\n--- 가우시안 분포 시각화 ---")

    # 평균 0, 표준편차 1인 표준 정규 분포에서 데이터 샘플링
    gaussian_data = np.random.randn(height, width)

    # 히스토그램 생성
    plt.figure(figsize=(10, 6))
    plt.hist(gaussian_data.flatten(), bins=bins, density=True, alpha=0.7, label='Sampled Gaussian Data')

    # 이론적인 확률 밀도 함수(PDF) 그리기
    x = np.linspace(-4, 4, 200)
    plt.plot(x, stats.norm.pdf(x, 0, 1), 'r-', lw=2, label='Ideal Gaussian PDF (N(0,1))')

    plt.title('Visualization of Gaussian (Normal) Distribution')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
visualize_gaussian_distribution()
```

## \#\# '평균 7' 템플릿과 노이즈 섞기

**진행 순서:**

1.  **MNIST 데이터셋 로드**: `torchvision` 라이브러리를 사용해 MNIST 손글씨 숫자 데이터셋을 불러옵니다.
2.  **'7' 데이터 필터링**: 전체 데이터셋에서 숫자 '7'에 해당하는 이미지들만 추출합니다.
3.  **'평균 7' 계산**: 추출된 모든 '7' 이미지들의 픽셀별 평균을 계산하여, 가장 평균적인 모습의 '7' 템플릿을 생성합니다.
4.  **노이즈와 혼합**: 이 '평균 7' 템플릿과 우리가 이전에 만든 가우시안 노이즈를 서로 다른 비율(`alpha` 값)로 섞어 결과를 확인합니다.

```python
import torch
import torchvision
import torchvision.transforms as transforms
import numpy as np
import matplotlib.pyplot as plt

# --- 1. MNIST 데이터셋 로드 ---
# 이미지를 텐서로 변환하는 Transform 정의
transform = transforms.ToTensor()
```

```python
transform
```

```python
# MNIST 학습 데이터셋 다운로드 및 로드
# 최초 실행 시 데이터 다운로드에 시간이 걸릴 수 있습니다.
train_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
```

```python
train_dataset
```

```python
# --- 2. '7' 데이터 필터링 ---
# 학습 데이터와 레이블(정답) 분리
data = train_dataset.data
targets = train_dataset.targets
```

```python
np.shape(data)
```

```python
np.shape(targets)
```

```python
targets[:10]
```

```python
# 레이블이 7인 데이터의 인덱스를 찾아 해당 데이터만 추출
sevens_data = data[targets == 7]
print(f"전체 학습 데이터 중 숫자 '7'의 개수: {len(sevens_data)}개")
```

```python
np.shape(sevens_data)
```

```python
np.mean(sevens_data.numpy()), np.std(sevens_data.numpy())
```

```python
# --- 3. '평균 7' 템플릿 계산 ---
# 평균 계산을 위해 데이터를 float 타입으로 변환하고 0~1 사이로 정규화
sevens_data_normalized = sevens_data.float() / 255.0
```

```python
np.mean(sevens_data_normalized.numpy()), np.std(sevens_data_normalized.numpy())
```

```python
# 모든 '7' 이미지에 대해 픽셀별 평균 계산
average_seven_template = torch.mean(sevens_data_normalized, dim=0)
print(f"계산된 '평균 7' 템플릿의 shape: {average_seven_template.shape}")
```

```python
# --- 4. 노이즈 생성 및 템플릿과 혼합 ---
# 28x28 가우시안 노이즈 생성
noise = np.random.randn(28, 28)
np.shape(noise)
```

```python
np.mean(noise), np.std(noise)
```

```python
# 혼합 비율(alpha) 설정
alphas = [0.3, 0.5, 0.7, 0.9, 0.95, 0.97]

# 시각화
fig, axes = plt.subplots(1, len(alphas) + 2, figsize=(15, 4))

# 원본 '평균 7' 템플릿 시각화
axes[0].imshow(average_seven_template.numpy(), cmap='gray')
axes[0].set_title("Average '7' Template")
axes[0].axis('off')

# 원본 노이즈 시각화
axes[1].imshow(noise, cmap='gray')
axes[1].set_title("Original Noise")
axes[1].axis('off')

# alpha 값에 따라 템플릿과 노이즈를 섞어 시각화
for i, alpha in enumerate(alphas):
    # blended_image = alpha * template + (1-alpha) * noise
    blended_image = alpha * average_seven_template.numpy() + (1 - alpha) * noise
    
    ax = axes[i + 2]
    ax.imshow(blended_image, cmap='gray')
    ax.set_title(f"Blended (alpha={alpha})")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

정규분포에 우리가 원하는 분포(현재 mnist 데이터 7모양)의 특징을 주입해 근사시킴.
